## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit

### The result should look something like this:

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


### The result should look something like this:
<img src="IMG/2.png" width=400>

```python


fig, axes = plt.subplots()

axes.plot(data.total_units, 'r', lw=3, ls='--', marker='o', markersize=6, markerfacecolor="blue")

axes.set_xlim([-1, 12])
axes.set_ylim([10000, 50000])

axes.set_xlabel("Month Number",labelpad=1)
axes.set_ylabel("Sold units number", labelpad=10)
axes.set_title("Company Sales data of last year");

axes.legend(["Profit Data of last year"], loc=4);

plt.show()
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).


### The result should look something like this:
<img src="IMG/3.png" width=400>

```python
fig, axes = plt.subplots()


axes.plot(data.facecream, 'b', lw=3, marker='o', markersize=6, markerfacecolor="blue")

axes.plot(data.facewash, 'y', lw=3, marker='o', markersize=6, markerfacecolor="yellow")
axes.plot(data.toothpaste, 'g', lw=3, marker='o', markersize=6, markerfacecolor="green")
axes.plot(data.bathingsoap, 'r', lw=3, marker='o', markersize=6, markerfacecolor="red")
axes.plot(data.shampoo, color="mediumpurple", lw=3, marker='o', markersize=6, markerfacecolor="mediumpurple")
axes.plot(data.moisturizer, color="brown", lw=3, marker='o', markersize=6, markerfacecolor="brown")


axes.set_xlabel("Month Number",labelpad=1)
axes.set_ylabel("Sold units in number", labelpad=10)
axes.set_title("Sales Data");

axes.legend(["Profit Data of last year"], loc=4)


axes.set_ylim([1000, 18000])

axes.legend(["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"])

plt.show()
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.


### The result should look something like this:
<img src="IMG/4.png" width=400>

```python
monate  = data ['month_number']
sales = data ['toothpaste']

plt.grid(True, linewidth= 0.5, linestyle="--")
plt.scatter(monate, sales, label="Zahnpastazahlen")
plt.xlabel("Month Number")
plt.ylabel("Number of units Sold")
plt.legend(loc=2)
plt.title('Tooth paste Sales data')
plt.ylim([4250, 8000])

plt.show()

```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.


### The result should look something like this:
<img src="IMG/5.png" width=400>

```python
plt.bar([i-0.25 for i in data["month_number"]], data["facecream"], label ="Face Cream sales data", align="edge", width= 0.2)
plt.bar([i+0.25 for i in data["month_number"]], data["facewash"], label = "Face Wash sales data", align="edge", width= 0.2)
plt.legend(loc=2)

plt.title("Gesichtscreme und Gesichtswaschcreme Verkaufszahlen")
plt.xlabel("Monat")
plt.ylabel("Verkaufszahlen")
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```
